```python
# HISTOGRAM EQULIZATION

import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img, cmap=None):
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```python
gorilla = cv2.imread('/home/srijanani/GIT/OpenCV/Images/gorilla.jpeg', 0)
```

```python
display(gorilla, cmap='gray')
```

```python
gorilla.shape
```

```python
hist_value = cv2.calcHist([gorilla], channels=[0], mask=None, histSize=[256], ranges=[0, 256])
```

```python
plt.plot(hist_value)
```

```python
# USING THE HISTOGRAM EQULIZATION FNCTION

eq_gorilla = cv2.equalizeHist(gorilla)
```

```python
display(eq_gorilla, cmap='gray')
```

```python
hist_value = cv2.calcHist([eq_gorilla], channels=[0], mask=None, histSize=[256], ranges=[0, 256])

plt.plot(hist_value)
plt.title('HISTOGRAM FOR EQUALIZATION GORILLA')
```
