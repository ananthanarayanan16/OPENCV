```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_1 = cv2.imread('/home/srijanani/GIT/OpenCV/Images/puppy.jpeg')
img_1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)

img_2 = cv2.imread('/home/srijanani/GIT/OpenCV/Images/home.jpeg')
img_2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB)

plt.imshow(img_1)
```

```python
plt.imshow(img_2)
```

```python
# BLENDING IMAGE WITH SAME SIZE
# RESIZING THE IMAGES USING THE RESIZE FUNCTION

img_1 = cv2.resize(img_1, (300, 300))

img_2 = cv2.resize(img_2, (300, 300))

img_1.shape
```

```python
img_2.shape
```

```python
# AFTER RESIZING THE FIRST IMAGE

plt.imshow(img_1)
```

```python
# AFTER RESIZING THE SECOND IMAGE

plt.imshow(img_2)
```

```python
# BLENDING THE IMAGE USING ADD WEIGHT FUNTION

blended = cv2.addWeighted(src1=img_1,alpha=1,src2=img_2,beta=1,gamma=1)

plt.imshow(blended)
```
