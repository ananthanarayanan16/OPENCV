```python
# CREATE A 5X5 ARRAY ALL ELEMENT IN THE ARRAY IS 10 AND CREATE ARRAY USING RANDOM.RAND() FUNCTION

import numpy as np

arr = np.ones(shape = (5,5))

arr = arr * 10

# print(arr)

type(arr)   # USED TO DISPLAY THE TYPE OF THE ARRAY

arr
```

```python
import numpy as np

np.random.seed(1)

arr = np.random.randint(low = 0, high = 100, size = (5,5))

arr
```

```python
# CREATE ARRAY USING RANDOM AND FIND MINIMUM ADN MAXIMUM VALUE IN THE ARRAY

import numpy as np

np.random.seed(1)

arr = np.random.randint(0,100,(5,5))

print("\nMinimum value of the array : ", arr.min())

print("\nMaximum value of the array : ", arr.max())

arr
```

```python
# CONVERT IMAGE INTO ARRAY USING NUMPY FUNCTIONS AND ASSIGN THE VALUE ZERO FOR EACH CHENNAL

import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from PIL import Image

pic = Image.open("/home/srijanani/GIT/OpenCV/Images/puppy.jpeg")

pic_array = np.asarray(pic)

pic_copy = pic_array.copy()

plt.imshow(pic_copy)
```
