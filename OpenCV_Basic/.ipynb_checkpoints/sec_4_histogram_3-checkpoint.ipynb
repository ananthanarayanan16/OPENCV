```python
# HISTOGRAM EQULIZATION

import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img, cmap=None):
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```python
display(gorilla, cmap='gray')
```
