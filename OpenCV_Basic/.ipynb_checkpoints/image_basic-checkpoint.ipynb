```python
#  PLAYING WITH RGB COLOR

import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from PIL import Image

pic = Image.open("/home/srijanani/GIT/OpenCV/Images/puppy.jpeg")

plt.imshow(pic)

pic_array = np.asarray(pic)

# print(pic_array)  // PRINT THE IMAGE AS ARRAY VALUES

pic_array.shape

# THIS FOR RED CHANNEL

plt.imshow(pic_array[:,:,0],cmap = 'gray')

# THIS FOR GREEN CHANNEL

plt.imshow(pic_array[:,:,1],cmap = 'gray')

# THIS FOR BLUE CHANNEL

plt.imshow(pic_array[:,:,2],cmap = 'gray')
```
