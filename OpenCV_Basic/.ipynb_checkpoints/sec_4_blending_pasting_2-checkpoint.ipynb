```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_1 = cv2.imread('/home/srijanani/GIT/OpenCV/Images/puppy.jpeg')
img_1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)

plt.imshow(img_1)
img_1.shape
```

```python
img_2 = cv2.imread('/home/srijanani/GIT/OpenCV/Images/home.jpeg')
img_2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB)

img_2 = cv2.resize(img_2, (150, 80))

plt.imshow(img_2)
```

```python
# GRABBING FROM THE IMAGE 2  (ROI)

x_start = 300 - 150
y_start = 168 - 80

roi = img_1[y_start : 386, x_start : 580]

plt.imshow(roi)
# roi.shape
```

```python
# CREATE MASK 

plt.imshow(img_2)
```

```python
# CONVERTING THE IMAGE INTO GRAY SCALE

img_2_gray = cv2.cvtColor(img_2, cv2.COLOR_RGB2GRAY)

plt.imshow(img_2_gray)
```

```python
# INCLUDE COLOR MAP FOR DISPLAY GRAY IMAGE

plt.imshow(img_2_gray, cmap = 'gray')
```

```python
# CONVERTING BALCK INTO WHITE FOR CREATE MASK

img_inv = cv2.bitwise_not(img_2_gray)

plt.imshow(img_inv)
```

```python
# USE COLOR MAP

plt.imshow(img_inv, cmap = 'gray')
```

```python
# CHECK SHAPE OF THE INVERSE IMAGE

img_inv.shape
```

```python
# USING NUMPY TO CREATE ARRAY

import numpy as np
```

```python
white_background = np.full(img_2.shape, 255, dtype = np.uint8)
```

```python
plt.imshow(white_background)
```

```python
bv = cv2.bitwise_or(white_background, white_background, mask = img_inv)

plt.imshow(bv)
```

```python
# GET THE FORE GROUND OF THE IMAGE 2

fg = cv2.bitwise_or(img_2, img_2, mask = img_inv)

# fg = cv2.resize(fg, (200, 200))

plt.imshow(fg)
```

```python
 fg.shape
```

```python
# FINAL ROI 

final_roi = cv2.bitwise_or(roi,fg)

plt.imshow(final_roi)
```

```python
large_img = img_1

small_img = final_roi

large_img[y_start:y_start+small_img.shape[0], x_start:x_start+small_img.shape[1]] = small_img

plt.imshow(large_img)
```
