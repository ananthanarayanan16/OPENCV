```python
#  SECTION 3 TASK
#  TASK-1
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import cv2
```

```python
img = cv2.imread("/home/srijanani/GIT/OpenCV/Images/puppy.jpeg")

plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
# TASK-2 FLIP IMAGE

flip_img = cv2.flip(img, 0)

plt.imshow(flip_img)
```

```python
# TASK-3 DRAW RECTANGLE ON THE DOG FACE

cv2.rectangle(img, (100,20), (200,120), (255,0,0), 1)

plt.imshow(img)
```

```python
# TASK-4 DRAW TRIANGLE ON THE IMAGE

img = cv2.imread("/home/srijanani/GIT/OpenCV/Images/puppy.jpeg")

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

vertices = np.array([[150,20], [100,100], [200,100]], dtype=np.int32)

vertices = vertices.reshape((-1,1,2))

cv2.polylines(img, [vertices], isClosed=True, color=(0,0,255), thickness=2)

plt.imshow(img)
```

```python
# FILL THE TRIANGLE 

cv2.fillPoly(img, [vertices], color = (0,0,255))

plt.imshow(img)
```

```python
#  DRAW CRICLE ON IMAGE USING MOUSE EVENT

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import cv2

#####FUNCTION#####

def draw_circle(event, x, y, flag, param):

    if event == cv2.EVENT_RBUTTONDOWN:

        cv2.circle(img,center=(x,y), radius=30, color=(0,0,255), thickness=2)

    elif event == cv2.EVENT_RBUTTONDOWN:

        cv2.circle(img, (x,y), 50, (0,0,255), 2)

cv2.namedWindow(winname='my_drawing')

cv2.setMouseCallback('my_drawing',draw_circle)

#####SHOW IMAGE WITH OPENCV#####

img = cv2.imread("/home/srijanani/GIT/OpenCV/Images/puppy.jpeg")

# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

while True:

    cv2.imshow('my_drawing',img)
    
    if cv2.waitKey(20) & 0xFF == 27:

        break

cv2.destroyAllWindows()
```
