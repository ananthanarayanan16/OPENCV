```python
# READING DOG FACE IMAGE

face = cv2.imread('/home/srijanani/GIT/OpenCV/Images/dog_face.jpg')

face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)

plt.imshow(face)
```

```python
# METHOD 2 : PASS ANY BUILDIN FUNCTION IN EVAL FUNCTION AS A 'STRING'

func = eval('sum')

func([1,2,3])
```

```python
# ALL THE 6 METHODS FOR COMPARISON IN A LIST
# NOTE HOW WE ARE USING STRINGS, LATER ON WE WILL USE THE EVAL FUNCTION() TO CONVERT THE FUNCTION

methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR', 'cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
my_method = eval('cv2.TM_CCOEFF')

res = cv2.matchTemplate(full, face, my_method)

plt.imshow(res)

min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(res)

print(min_val, max_val, min_loc, max_loc)
```

```python
for m in methods:

    # CREATE A COPY

    full_copy = full.copy()

    method = eval(m)

    # TEMPLATE MATCHING

    res = cv2.matchTemplate(full_copy, face, method)

    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)

    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc

    else:
        top_left = max_loc

    height, width, channel = face.shape

    bottom_right = (top_left[0]+width, top_left[1]+height)

    cv2.rectangle(full_copy, top_left, bottom_right, (255, 0, 0), 10)

    # PLOT AND SHOW THE IMAGES

    plt.subplot(121)
    plt.imshow(res)
    plt.title('HEAPMAP OF TEMPLATE MATCHING')

    plt.subplot(122)
    plt.imshow(full_copy)
    plt.title('DETECTION OF TEMPLATE')

    # TITLE WITH THE METHOD USED
    plt.suptitle(m)

    plt.show()

    print('\n')
    print('\n')
```
