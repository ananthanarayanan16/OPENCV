```python
# TEMPLATE MATCHING

import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
full = cv2.imread('/home/srijanani/GIT/OpenCV/Images/dog.jpg')
full = cv2.cvtColor(full, cv2.COLOR_BGR2RGB)

plt.imshow(full)
```

```python
face = cv2.imread('/home/srijanani/GIT/OpenCV/Images/dog_face.jpg')
face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)

plt.imshow(face)
```

```python
# TEMPLATE MATCHING USE THE SUM OF SQUARE DIFFERENCE METHOD() THIS FUNCTION LOW VALUE IS PERFECT MATCH

result = cv2.matchTemplate(full, face, cv2.TM_SQDIFF)

plt.imshow(result)
plt.title('HEATMAP FOR TEMPLATE MATCHING')
# plt.imshow(result, cmap='gray')
```

```python
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

print(min_val, max_val, min_loc, max_loc)
```

```python
face.shape

height, width = face.shape[:2]

print(height, width)
```

```python
top_left = min_loc

bottom_right = top_left[0] + height, top_left[1] + width

full_copy = full.copy()

cv2.rectangle(full_copy, top_left, bottom_right, (0, 0, 255), 10)

plt.imshow(full_copy)
plt.title('DETECTION OF TEMPLATE')
```

```python
# TEMPLAGE USE THE SUM OF SQUARE DIFFERENCE NORMALIZED METHOD() THIS LOW VALUE IS PERFECT MATCHING

full_copy = full.copy()

result = cv2.matchTemplate(full_copy, face, cv2.TM_SQDIFF_NORMED)

plt.imshow(result, cmap='gray')
plt.title('Template Matching Result')
plt.colorbar()
plt.show()
```

```python
# GET THE MINIMUM AND MAXIMUM VALUE AND THEIR CORRESPONDING LOCATION

min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

print(min_val, max_val, min_loc, max_loc)
```

```python
height, width = face.shape[:2]

print(height, width)
```

```python
# POINT DECLERE FOR DRAW RECTANGLE

top_left = min_loc

bottom_right = top_left[0] + height, top_left[1] + width 
```

```python
# DRAW RECTANGLE ON THE FULL IMAGE

cv2.rectangle(full_copy, top_left, bottom_right, (0, 255, 0), 10)

plt.imshow(full_copy)
plt.title('DETECTION OF TEMPLATE')
# plt.colorbar()
# plt.show()
```
