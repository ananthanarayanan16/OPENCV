```python
# CANNY EDGE DETECTION

import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('/home/srijanani/GIT/OpenCV/Images/dog_face.jpg')

plt.imshow(img)
```

```python
edge = cv2.Canny(image=img, threshold1=127, threshold2=127)

plt.imshow(edge)
```

```python
edge = cv2.Canny(image=img, threshold1=0, threshold2=255)

plt.imshow(edge)
```
