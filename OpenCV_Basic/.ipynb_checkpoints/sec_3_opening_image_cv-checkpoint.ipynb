```python
# MATPLOTLIB --> RGB RED GREEN BLUE
# OPENCV --> BGR BLUE GREEN RED

# CONVERT BGR TO RGB 

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import cv2   #  IMPORT CV2 FILE HERE

img = cv2.imread("/home/srijanani/GIT/OpenCV/Images/puppy.jpeg")

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
# READING IMAGE IN GRAYSCALE

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import cv2

img = cv2.imread('/home/srijanani/GIT/OpenCV/Images/puppy.jpeg',cv2.IMREAD_GRAYSCALE)

plt.imshow(img)

img.shape    # GRAYSCALE IMAGE IS 2D IMAGE

plt.imshow(img, cmap = 'gray')
```
