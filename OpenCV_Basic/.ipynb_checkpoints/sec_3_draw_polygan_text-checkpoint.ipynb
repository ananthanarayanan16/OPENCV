```python
# DRAW POLYGAN AND TEXT IN THE IMAGE

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import cv2
```

```python
blank_img = np.zeros((512,512,3), dtype = np.uint16)

plt.imshow(blank_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX

cv2.putText(blank_img, text='SrinivasH', org=(50,450), fontFace=font, fontScale=3, color=(255,255,255),
           thickness=3, lineType=cv2.LINE_8)

plt.imshow(blank_img)
```

```python
# CREATE ANOTHER BLANK IMAGE

blank_img = np.zeros(shape=(512,512,3), dtype=np.int32)

plt.imshow(blank_img)
```

```python
# CREATE VERTICES FOR POLYGON 

vertices = np.array([[100,300],[200,200], [400,300], [200,400]], dtype=np.int32)

vertices.shape
```

```python
#  CONVERTING 2D TO 3D 

pts = vertices.reshape(-1,1,2)

pts.shape
```

```python
cv2.polylines(blank_img, [pts], isClosed=True, color=(255,0,0), thickness=5)

plt.imshow(blank_img)
```
