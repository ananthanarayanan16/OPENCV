```python
# DRAW BASIC SHAPE ON THE IMAGE

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import cv2

# blank_img = np.zeros(shape = (512,512,3), dtype = np.int16)

# plt.imshow(blank_img)
```

```python

blank_img = np.zeros(shape = (512,512,3), dtype = np.int16)

plt.imshow(blank_img)
```

```python
 cv2.rectangle(blank_img, pt1 = (350,20), pt2 = (500,200), color = (0,255,0), thickness = 10)

plt.imshow(blank_img)
```

```python
# DRAW ANOTHER RECTANGLE 

cv2.rectangle(blank_img, pt1=(150,250), pt2=(250,350), color=(255,0,0), thickness = 10)

plt.imshow(blank_img)
```

```python
cv2.circle(blank_img, center=(100,100), radius=50, color=(0,0,255), thickness=5)

plt.imshow(blank_img)
```

```python
# FILLING INSIDE THE RECTANGLE

cv2.rectangle(blank_img, pt1=(50,375), pt2=(150,475), color=(120,200,210), thickness=-1)

plt.imshow(blank_img)
```

```python
# FILLING INSIDE THE CIRCLE 

cv2.circle(blank_img, center=(400,400), radius=50, color=(123,134,145), thickness=-1)

plt.imshow(blank_img)
```

```python
# DRAW LINE IN THE IMAGE

cv2.line(blank_img, pt1=(0,0), pt2=(512,512),color=(130,199,120), thickness=5)

plt.imshow(blank_img)
```
