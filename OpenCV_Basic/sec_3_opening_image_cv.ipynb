```python
#  RESIZING THE IMAGE USING OPENCV FUNCTION

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import cv2

img = cv2.imread('/home/srijanani/GIT/OpenCV/Images/puppy.jpeg')

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

print("\nOriginal image size : ",img.shape)

resize_img = cv2.resize(img, (200, 120))

print("\nAfter resizing the image size : ",resize_img.shape)

# RESIZING USING RATIO

width_ratio = 0.5

height_ratio = 0.5

ratio_resize_img = cv2.resize(img, (0,0), img, width_ratio, height_ratio)

print("\nAfter resize using the ratio parameter : ",ratio_resize_img.shape)

plt.imshow(ratio_resize_img)
```

```python
# FLIPING IMAGE

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import cv2

img = cv2.imread('/home/srijanani/GIT/OpenCV/Images/puppy.jpeg')

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#  FLIP THE IMAGE USING THE FLIP() FUNCTION

flip_img = cv2.flip(img, -)

plt.imshow(flip_img)
```

```python
# WRITE IMAGE 

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import cv2

img = cv2.imread('/home/srijanani/GIT/OpenCV/Images/puppy.jpeg')  # READ IMAGE USING IMREAD FUNCTION

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)

# WRITE IMAGE USING THE IMWRITE() FUNCTION

cv2.imwrite('duplicate_puppy.jpeg', img)  

# THE WRITE IMAGE IS STORE. WHERE THE CODE FILE IS LOCATED THAT PLACE THE IMAGE FILE IS SAVE

fig = plt.figure(figsize = (10,10))

ax = fig.add_subplot(111)

ax.imshow(img)
```
