```python
# GRADIENTS

import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_img(img):
    fig = plt.figure(figsize = (12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = cv2.imread('/home/srijanani/GIT/OpenCV/Images/sudoku.jpeg', 0)
```

```python
display_img(img)
```

```python
# SOBEL OPERATOR USING X GRADIENT

sobel_x = cv2.Sobel(img, cv2.CV_64F,1,0,ksize=5)

display_img(sobel_x)

# plt.imshow(sobelx)
```

```python
# SOBEL OPERATOR USING Y GRADIENT

sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)

display_img(sobel_y)
```

```python
# USIING LAPLACIAN OPERATOR

laplacian = cv2.Laplacian(img, cv2.CV_64F)

display_img(laplacian)
```

```python
# BLENDING TWO IMAGES

blend = cv2.addWeighted(src1=sobel_x, alpha=0.5, src2=sobel_y, beta=0.5, gamma=0)

display_img(blend)
```

```python
# USING THRESHOLDING 

ret, threshold = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

display_img(threshold)
```

```python
kernal = np.ones((4,4), dtype=np.uint8)

gradient = cv2.morphologyEx(blend, cv2.MORPH_GRADIENT, kernal)

display_img(gradient)
```
