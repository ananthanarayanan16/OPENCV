```python
# GRID DETECTION

import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('/home/srijanani/GIT/OpenCV/Images/flat_chessboard.jpg')
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)

plt.imshow(flat_chess)
```

```python
# FIND CORNER USING THE FUNCTION FINDCHESSBOARDCORNERS METHOD()

found , corners = cv2.findChessboardCorners(flat_chess, (7, 7))
```

```python
found
```

```python
# DRAW DETECTIN CORNER USE DRAWCHESSBOARD CORNER METHOD()

cv2.drawChessboardCorners(flat_chess, (7, 7), corners, found)
```

```python
plt.imshow(flat_chess)
```
