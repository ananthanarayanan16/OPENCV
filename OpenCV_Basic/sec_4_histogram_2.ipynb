```python
# USING THE HISTOGRAM EQULIZERS

import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rainbow = cv2.imread('/home/srijanani/GIT/OpenCV/Images/rainbow.jpeg') # ORIGINAL COLOR BGR
show_rainbow = cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB) # CONVERT COLOR RGB
```

```python
plt.imshow(show_rainbow)
```

```python
img = rainbow
```

```python
img.shape
```

```python
img.shape[:2]
```

```python
# CREATING MASK

mask = np.zeros(img.shape[:2], np.uint8)
```

```python
plt.imshow(mask, cmap='gray')
```

```python
# CREATE WHITE COLOR RETANGLE IN THE MASK IMAGE

mask[2500:3100, 200:2800] = 255  # ASSIGNING THE VALUE 255 IN THE SILICE AREA OF THE IMAGE

plt.imshow(mask, cmap='gray')
```

```python
# plt.imshow(show_rainbow)
```

```python
masked_img = cv2.bitwise_or(img, img, mask=mask)
```

```python
plt.imshow(masked_img)
```

```python
show_mask_img = cv2.bitwise_and(show_rainbow, show_rainbow, mask = mask)
```

```python
plt.imshow(show_mask_img)
```

```python
# OPENCV DEFAULT B G R 

hist_mask_value_red = cv2.calcHist([rainbow], channels=[2], mask=mask, histSize=[256], ranges=[0, 256])
```

```python
plt.plot(hist_mask_value_red)
plt.title('RED HISTOGRAM FOR MASKED RAINBOW')
```

```python
hist_value_red = cv2.calcHist([rainbow], channels=[2], mask=None, histSize=[256], ranges=[0, 256])
```

```python
plt.plot(hist_value_red)
plt.title('RED HISTOGRAM FOR NORAML RAINBOW ')
```
