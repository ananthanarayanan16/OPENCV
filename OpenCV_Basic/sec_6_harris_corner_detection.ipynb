```python
# FIND CORNERS IN IMAGE USING HARRIS CORNER DETETION

import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# READING THE FALT CHESS BOARD IMAGE

flat_chess = cv2.imread('/home/srijanani/GIT/OpenCV/Images/flat_chessboard.jpg')
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)

plt.imshow(flat_chess)
```

```python
# CONVERT COLOR INTO GRAY SCALE IMAGE

gray_flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_RGB2GRAY)

plt.imshow(gray_flat_chess, cmap='gray')
```

```python
# READING REAL CHESS BOARD IMAGE

real_chess = cv2.imread('/home/srijanani/GIT/OpenCV/Images/chessboard.jpg')
real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2RGB)

plt.imshow(real_chess)
```

```python
# CONVERTING COLOR INTO GRAY SCALE IMAGE

gray_real_chess = cv2.cvtColor(real_chess, cv2.COLOR_RGB2GRAY)

plt.imshow(gray_real_chess, cmap='gray')
```

```python
gray_flat_chess
```

```python
# THE GRAY FLAT CHESS BOARD IN DATA TYPE 'INT'. IT CONVERT INTO 'FLOAT' VALUES FOR USING CORNER DETECTION

gray = np.float32(gray_flat_chess)

gray  # THE GRAY_FLAT_CHESSBOARD CONVERT INTO FLOAT
```

```python
# CORNER IN FLAT CHESSBOARD IMAGE

dst = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.04)
```

```python
dst = cv2.dilate(dst, None)
```

```python
dst.max()
```

```python
flat_chess[dst > 0.01 * dst.max()] = [255, 0, 0]

plt.imshow(flat_chess)
```

```python
# CORNER DETECTION USING THE REAL CHESS BOARD IMAGE

gray = np.float32(gray_real_chess)  # CONVERTING INT INTO FLOAT THE REAL CHESS BOARD IMAGE

gray
```

```python
dst = cv2.cornerHarris(src=gray, blockSize=5, ksize=5, k=0.04)
```

```python
dst = cv2.dilate(dst, None)
```

```python
real_chess[dst>0.01 * dst.max()] = [255, 0, 0] 
```

```python
plt.imshow(real_chess)
```
