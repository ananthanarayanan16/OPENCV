```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# TASK DISPLAY THE GIRAFFE IMAGE

def display_image(img, cmap=None):
    fig = plt.figure(figsize=(8, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```python
img = cv2.imread('/home/srijanani/GIT/OpenCV/Images/giraffe.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
display_image(img)

# plt.imshow(img)  USING MATPLOTLIB
```

```python
# SECOND TASK APPLY BINARY THRESHOLD ONTO THE IMAGE

# THE GIRAFFE IMAGE CONVERT INTO GRAY SCALE IMAGE

gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')  # DISPLAY THE GRAY SCALE IMAGE
```

```python
# APPLYING THRESHOLD BINARY ONTO THE GRAY IMAGE

res, threshold = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

display_image(threshold, cmap='gray')
```

```python
# TASK THREE CONVERT THE GIRAFFE COLOR IMAGE INTO HSV COLOR IMAGE AND DISPLAY IT

img = cv2.imread('/home/srijanani/GIT/OpenCV/Images/giraffe.jpg')

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # BGR COLOR IMAGE CONVERT INTO RGB COLOR IMAGE

plt.imshow(img)
```

```python
hsv_img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

display_image(hsv_img)
```
